```python
#–ó–∞–¥–∞—á–∞ 1
def MaxScore():
            inputArray = []
            n =int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏:"))
            if ((n<1) or (n>105)):
                return "–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤"
            for i in range(n):
                f = int(input("–í–≤–µ–¥–∏—Ç–µ —ç–ª–µ–º–µ–Ω—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏: ")) 
                if ((f<1) or (f>105)):
                    return("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–∞")
                inputArray.append(f)
            
            a = inputArray
            a.sort(reverse=True)

            globalSum = 0
            currentNumber = None

            sum1 = a[0]
            firstNumber = a[0]

            sum2 = None
            secondNumber = None
            for i in range (len(a)):
                if (currentNumber != None):
                    if(abs(currentNumber - a[i]) <= 1):
                        continue

                if (firstNumber==None):
                    firstNumber = a[i]

                if (a[i] == firstNumber):
                    sum1 += a[i]
                elif (secondNumber==None):
                    secondNumber = a[i]
                    sum2 = a[i]
                elif (secondNumber == a[i]):
                    sum2 += a[i]
                else:
                    if (sum1 > sum2):
                        globalSum = sum1
                        currentNumber = firstNumber
                    else:
                        globalSum = sum2
                        currentNumber = secondNumber
                    sum1 = 0
                    sum2 = None
                    firstNumber = None
                    secondNumber = None

                    if (abs(currentNumber - a[i]) > 1):
                        sum1 = a[i]                     
                        firstNumber = a[i]                       
                        
            if (sum2!=None):
                if(sum2>sum1):
                    globalSum+=sum2
                else:
                    globalSum+=sum1
            return ("–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—á–∫–æ–≤:", globalSum)

print(MaxScore())
print()
print(MaxScore())
print()
print(MaxScore())
```

```python
# –ó–∞–¥–∞—á–∞ 2
import re
def substring():
    s = input("–í–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫—É: ")
    s = s.upper()
    if (len(s)<1 or len(s)>105):
        return ("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω–∞—è –¥–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏")
    if re.search("(AB.*BA|BA.*AB)", s):
        return("YES")
    else: return("NO")

print(substring())
print()
print(substring())
print()
print(substring())
```

```python
#–ó–∞–¥–∞—á–∞ 4

def Eight():
    s = input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ: ")
    if (len(s)>100):
        return ("–°–ª–∏—à–∫–æ–º –±–æ–ª—å—à–æ–µ —á–∏—Å–ª–æ")
    if (s[0]=='0'):
        return ("–ß–∏—Å–ª–æ —Å–æ–¥–µ—Ä–∂–∏—Ç –≤–µ–¥—É—â–∏–µ –Ω–µ–∑–Ω–∞—á–∞—â–∏–µ –Ω—É–ª–∏")
    for i in range (len(s)):
        if(s[i]==0):
            return("yes: 0")
        t1 = int(s[i])
        if (t1 % 8 == 0):
            return("yes: ", t1) 
        for j in range((i+1),len(s)): 
            t2 = t1 * 10 + int(s[j])
            if (t2 % 8 == 0):
                return("yes: ", t2)
            for k in range ((j+1),len(s)):
                t3 = t2 * 10 + int(s[k])
                if (t3 % 8 == 0):
                    return("yes: ", t3)
    return("no")

print(Eight())
print()
print(Eight())
print()
print(Eight())
print()
print(Eight())
print()
print(Eight())
```
